## Filtros

```python
def gausian_blur(img, order=21, sigma=21):
    np_img = np.array(img)
    filtered_img = cv2.GaussianBlur(np_img, (order,order), sigma)
    return Image.fromarray(filtered_img)

def bilateral_filter(img):
    """
    d: Diameter of each pixel neighborhood.
    
    sigmaColor: Value of \sigma in the color space. The greater the value, the colors farther to each other will start to get mixed.
    
    sigmaSpace: Value of \sigma in the coordinate space. The greater its value, the more further pixels will mix together, given that
    their colors lie within the sigmaColor range.    
    """
    np_img = np.array(img)
    filtered_img = cv2.bilateralFilter(np_img, 15, 75, 75)
    return Image.fromarray(filtered_img) 

def canny_edge_filter(img, min_val=40, max_val=85):
    img = np.array(img)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) # PIL (RGB) -> OpenCV (BGR)
    img = cv2.Canny(img, min_val, max_val)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # OpenCV (BGR) -> PIL (RGB)
    return Image.fromarray(img)
```

## Creaci칩n de los dataframes

```python
transform_r = transforms.Compose \
(
    [
        transforms.Resize((IMAGE_RESIZE, IMAGE_RESIZE)),
        transforms.Grayscale(num_output_channels=1),
        transforms.ToTensor(),
        transforms.Normalize((IMAGE_NORMALIZE,), (IMAGE_NORMALIZE,))
    ]
)
transform_b = transforms.Compose \
(
    [
        transforms.Resize((IMAGE_RESIZE, IMAGE_RESIZE)),
        transforms.Lambda(lambda img: bilateral_filter(img)),
        transforms.Grayscale(num_output_channels=1),
        transforms.ToTensor(),
        transforms.Normalize((IMAGE_NORMALIZE,), (IMAGE_NORMALIZE,))
    ]
)
transform_c = transforms.Compose \
(
    [
        transforms.Resize((IMAGE_RESIZE, IMAGE_RESIZE)),
        transforms.Lambda(lambda img: canny_edge_filter(img)),
        transforms.Grayscale(num_output_channels=1),
        transforms.ToTensor(),
        transforms.Normalize((IMAGE_NORMALIZE,), (IMAGE_NORMALIZE,))
    ]
)

def get_sub_data_loader(dataset, fraction_of_data=0.1):
    dataset_size = len(dataset)
    indices = list(range(dataset_size))
    subset_size = int(np.floor(fraction_of_data * dataset_size))
    np.random.shuffle(indices)
    subset_indices = indices[:subset_size]
    sub_dataset = Subset(dataset, subset_indices)
    # print(f'Subset size: {len(sub_dataset)}')
    return DataLoader(dataset=sub_dataset, batch_size=BATCH_SIZE, shuffle=True)

# Train data
# Raw
train_dataset_r = datasets.ImageFolder(root=TRAIN_DATA_PATH, transform=transform_r)
train_loader_r = DataLoader(dataset=train_dataset_r, batch_size=BATCH_SIZE, shuffle=True)
train_sub_loader_r = get_sub_data_loader(dataset=train_dataset_r)
# Bilateral
train_dataset_b = datasets.ImageFolder(root=TRAIN_DATA_PATH, transform=transform_b)
train_loader_b = DataLoader(dataset=train_dataset_b, batch_size=BATCH_SIZE, shuffle=True)
train_sub_loader_b = get_sub_data_loader(dataset=train_dataset_b)
# Canny
train_dataset_c = datasets.ImageFolder(root=TRAIN_DATA_PATH, transform=transform_c)
train_loader_c = DataLoader(dataset=train_dataset_c, batch_size=BATCH_SIZE, shuffle=True)
train_sub_loader_c = get_sub_data_loader(dataset=train_dataset_c)

# Test data
# Raw
test_dataset_r = datasets.ImageFolder(root=TEST_DATA_PATH, transform=transform_r)
test_loader_r = DataLoader(dataset=test_dataset_r, batch_size=BATCH_SIZE, shuffle=True)
# Bilateral
test_dataset_b = datasets.ImageFolder(root=TEST_DATA_PATH, transform=transform_b)
test_loader_b = DataLoader(dataset=test_dataset_b, batch_size=BATCH_SIZE, shuffle=True)
# Canny
test_dataset_c = datasets.ImageFolder(root=TEST_DATA_PATH, transform=transform_c)
test_loader_c = DataLoader(dataset=test_dataset_c, batch_size=BATCH_SIZE, shuffle=True)


print(f'Tama침o del conjunto de entrenamiento: {len(train_dataset_r)}')
print(f'Tama침o del conjunto de testing: {len(test_dataset_r)}')
```

## Funci칩n de entrenamiento

```python
def train(model, train_data_loader, model_name):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2, verbose=True)

    for epoch in range(EPOCHS):
        
        # Para llevar un registro de la p칠rdida por epoch
        epoch_loss = 0

        for images, labels in train_data_loader:

            # Limpiar gradientes: Se restablecen los gradientes acumulados en el optimizador a cero para evitar que se sumen a los gradientes del lote actual.
            optimizer.zero_grad()

            # Forward pass: Se pasan las im치genes a trav칠s del modelo para obtener las predicciones (model_predictions).
            model_predictions = model(images)

            # print("Model predictions shape:", model_predictions.shape)
            # print("Labels shape:", labels.shape)

            # C치lculo de la p칠rdida: Se calcula la p칠rdida utilizando la funci칩n de p칠rdida definida anteriormente,
            # comparando las predicciones (model_predictions) con las etiquetas verdaderas (labels).
            loss = criterion(model_predictions, labels)
            
            # Sumar la p칠rdida de este lote a la p칠rdida total de la 칠poca
            epoch_loss += loss.item()

            # Backward pass: Se realiza la propagaci칩n hacia atr치s para calcular los gradientes de la p칠rdida con respecto a los par치metros del modelo.
            loss.backward()

            # Actualizar par치metros: Se actualizan los par치metros del modelo utilizando los gradientes calculados en la etapa anterior.
            optimizer.step()

        # Imprimir p칠rdida promedio por 칠poca y el LR
        avg_loss = epoch_loss / len(train_data_loader)
        current_lr = optimizer.param_groups[0]['lr']  # Obtener el learning rate actual
        print(f'Epoch [{epoch + 1}/{EPOCHS}], Loss: {avg_loss:.4f}, Learning Rate: {current_lr:.6f}')
        # print(f'Epoch [{epoch + 1}/{EPOCHS}], Loss: {loss.item():.4f}')

        # Actualiza el scheduler con la p칠rdida promedio
        scheduler.step(avg_loss)

        wandb.log({
            f"{model_name}_epoch_loss": avg_loss
        }, commit=True)
```

- **mode**='min': El scheduler se activar치 si la p칠rdida se encuentra en su m칤nimo.
- **factor**=0.5: Reducir치 el learning rate a la mitad.
- **patience**=2: Esperar치 2 칠pocas antes de reducir el learning rate si no hay mejora.

## Funciones de evaluaci칩n

```python
def evaluate_model(model, test_data_loader):
    correct = 0
    total = 0

    with torch.no_grad():

        for images, labels in test_data_loader:

            model_predictions = model(images)
            _, predicted = torch.max(model_predictions.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    print(f'Accuracy of the model on the test images: {100 * correct / total:.2f}%')


def get_predictions(model, data_loader):
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in data_loader:
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    return np.array(all_labels), np.array(all_preds)


def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    
    disp.plot(cmap=plt.cm.Blues)
    plt.title('Confusion matrix')
    plt.show()
```

#### Modelo 1

```python
class CNN1(nn.Module):
    def __init__(self):
        super(CNN1, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.inception1 = self.InceptionBlock(in_channels=32) # -> 128
        self.conv2 = nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, stride=1, padding=1)

        self.fc1 = nn.Linear(in_features=(128 * 16 * 16), out_features=64)
        self.fc2 = nn.Linear(in_features=64, out_features=3)

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dropout = nn.Dropout(DROPOUT_RATE)

    class InceptionBlock(nn.Module):
        def __init__(self, in_channels):
            super(CNN1.InceptionBlock, self).__init__()
            self.branch1x1 = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)

            self.branch3x3_1 = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)
            self.branch3x3_2 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=3, padding=1)

            self.branch5x5_1 = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)
            self.branch5x5_2 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=5, padding=2)

            self.branch_pool = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)

        def forward(self, x):
            branch1 = self.branch1x1(x)

            branch2 = self.branch3x3_1(x)
            branch2 = self.branch3x3_2(branch2)

            branch3 = self.branch5x5_1(x)
            branch3 = self.branch5x5_2(branch3)

            branch4 = F.max_pool2d(x, kernel_size=3, stride=1, padding=1)
            branch4 = self.branch_pool(branch4)

            outputs = [branch1, branch2, branch3, branch4]
            return torch.cat(outputs, 1) # (32 + 32 + 32 + 32 = 128)

    def forward(self, x):
        # Etapa 1
        x = self.pool(F.relu(self.conv1(x)))    # [32, 1, 128, 128] -conv-> [32, 32, 128, 128] -ReLu-> same -pooling-> [32, 32, 64, 64]
        x = self.pool(self.inception1(x))       # [32, 32, 64, 64] -incep-> [32, 128, 64, 64] -pooling-> [32, 128, 32, 32]
        x = self.pool(F.relu(self.conv2(x)))    # [32, 128, 32, 32] -conv-> [32, 128, 32, 32] -ReLu-> same -pooling-> [32, 128, 16, 16]
        # Etapa 2
        x = x.view(-1, 128 * 16 * 16)           # [32, 128, 16, 16] -view-> [32, 128*16*16]
        # Etapa 3
        x = F.relu(self.fc1(x))                 # [32, 128*16*16] -fc-> [32, 64]
        x = self.dropout(x)                     # Aplicar Dropout
        x = self.fc2(x)                         # [32, 64] -fc-> [32, 3]
        return x
```

#### Justificaci칩n

La arquitectura propuesta para esta CNN est치 dise침ada de manera que logre un equilibrio entre simplicidad y capacidad de extracci칩n de caracter칤sticas en im치genes. A continuaci칩n, justifico las decisiones tomadas en la estructura del modelo:

##### 1. **Capas convolucionales (Conv2D)**
Se han utilizado tres capas convolucionales con un n칰mero creciente de filtros (32, 128) y un tama침o de kernel de 3x3, junto con un bloque Inception. Esta configuraci칩n permite:
- **Extracci칩n progresiva de caracter칤sticas**: Al aumentar el n칰mero de canales, la red puede capturar detalles de baja a alta complejidad a medida que las im치genes pasan por las capas.
- **Receptive field**: El tama침o de 3x3 es un est치ndar eficaz que equilibra precisi칩n y costo computacional, permitiendo que la red capture patrones locales en las im치genes.

##### 2. **Bloque Inception**
La incorporaci칩n de un bloque Inception permite que la red capture una variedad m치s amplia de caracter칤sticas a diferentes escalas. Las ventajas incluyen:
- **M칰ltiples tama침os de convoluci칩n**: El bloque utiliza convoluciones con diferentes tama침os de kernel (1x1, 3x3, 5x5) en paralelo, permitiendo que la red aprenda a detectar caracter칤sticas en diversas escalas.
- **Reducci칩n de dimensionalidad**: Las convoluciones de 1x1 ayudan a reducir el n칰mero de canales, optimizando la carga computacional.

##### 3. **Max Pooling**
El uso de `MaxPool2D` con kernel 2x2 reduce las dimensiones espaciales de las im치genes (de 128x128 a 16x16 tras 3 pasos), permitiendo:
- **Reducci칩n de la dimensionalidad**: Lo cual es clave para disminuir el n칰mero de par치metros en las capas completamente conectadas (fully connected), evitando el sobreajuste.
- **Retenci칩n de informaci칩n importante**: El max pooling retiene los valores m치ximos en cada regi칩n, preservando caracter칤sticas clave mientras descarta informaci칩n redundante.

##### 4. **Capas completamente conectadas (Fully Connected)**
Se ha elegido un n칰mero moderado de neuronas en las capas `fc1` (64) y `fc2` (3), con las siguientes razones:
- **Compactaci칩n de las caracter칤sticas**: La salida de la 칰ltima capa convolucional se transforma en un vector de caracter칤sticas mediante `view`, y las capas fully connected resumen esta informaci칩n en un formato 칰til para la clasificaci칩n.
- **Reducci칩n final a 3 clases**: La capa `fc2` genera una salida de 3 neuronas, correspondiente a las 3 categor칤as del problema.

##### 5. **Dropout**
Se introduce dropout en las 칰ltimas iteraciones con el fin de:
- **Evitar el sobreajuste**: Dropout es una t칠cnica eficaz para mejorar la generalizaci칩n del modelo, ya que previene la dependencia excesiva en ciertos patrones durante el entrenamiento.

##### 6. **Resultados experimentales**
El desempe침o del modelo mejora notablemente a lo largo de las iteraciones al ajustar los hiperpar치metros (learning rate, epochs) y al incluir dropout. Esto valida que la arquitectura es capaz de aprender representaciones 칰tiles cuando se le da suficiente tiempo de entrenamiento y se aplican t칠cnicas de regularizaci칩n adecuadas.

##### 7. **Etapas**
Las **etapas** de esta red CNN se pueden dividir en tres bloques principales:

###### 1. **Etapa de extracci칩n de caracter칤sticas**
   - **Capas**: `conv1`, `inception1`, `conv2`, y `pool`
   - **Descripci칩n**: Esta etapa utiliza capas convolucionales y de pooling para extraer caracter칤sticas de la imagen de entrada. La activaci칩n `ReLU` a침ade no linealidad y el `MaxPool2D` reduce la dimensionalidad de los mapas de caracter칤sticas.
     - Conv1: Extrae caracter칤sticas iniciales de la imagen (bordes, texturas).
     - Inception1: Captura caracter칤sticas en m칰ltiples escalas.
     - Conv2: Extrae a칰n m치s caracter칤sticas complejas.
     - MaxPool: Reduce la dimensi칩n espacial despu칠s de cada convoluci칩n.

###### 2. **Etapa de aplanamiento (Flattening)**
   - **Capas**: `x.view(-1, 128 * 16 * 16)`
   - **Descripci칩n**: Aplana la salida de la 칰ltima capa convolucional en un vector para poder pasarla a las capas totalmente conectadas.

###### 3. **Etapa de clasificaci칩n**
   - **Capas**: `fc1`, `dropout`, `fc2`
   - **Descripci칩n**: Las capas completamente conectadas (`fc1` y `fc2`) se encargan de la clasificaci칩n final. `fc1` reduce la dimensionalidad a 64 neuronas, y despu칠s se aplica un `dropout` para prevenir el sobreajuste. Finalmente, `fc2` produce la salida de 3 clases (asumiendo una clasificaci칩n de 3 categor칤as).

Estas tres etapas forman la estructura completa de la CNN.

##### Transformaci칩n de `x` a trav칠s de la red
El tensor de entrada `x` tiene la forma `[N, C, H, W]`, donde:
- **N** = Batch size (n칰mero de muestras).
- **C** = N칰mero de canales.
- **H** = Altura de la entrada.
- **W** = Ancho de la entrada.

- **Input shape**: `[32, 1, 128, 128]`
1. `[32, 1, 128, 128]`  **Conv1**  `[32, 32, 128, 128]`  **ReLU**  `[32, 32, 128, 128]`  **Pooling**  `[32, 32, 64, 64]`
2. `[32, 32, 64, 64]`  **Inception1**  `[32, 128, 64, 64]`  **Pooling**  `[32, 128, 32, 32]`
3. `[32, 128, 32, 32]`  **Conv2**  `[32, 128, 32, 32]`  **ReLU**  `[32, 128, 32, 32]`  **Pooling**  `[32, 128, 16, 16]`
4. `[32, 128, 16, 16]`  **View**  `[32, 128 * 16 * 16]`
5. `[32, 128 * 16 * 16]`  **Fully Connected (fc1)**  `[32, 64]`
6. `[32, 64]`  **Fully Connected (fc2)**  `[32, 3]`

###### `x.view()` function:
La funci칩n `view()` en PyTorch se utiliza para cambiar la forma de un tensor sin alterar sus datos subyacentes.
Transforma el tensor de entrada de una forma 4D `[N, C, H, W]` a una forma 2D `[N, F]`, donde:
- **F** = N칰mero de caracter칤sticas (features).
El n칰mero de caracter칤sticas se calcula como:
\[
F = C \times H \times W
\]

#### Explicaci칩n de las capas

##### Capas de Pooling
Una capa de pooling reduce la dimensionalidad del mapa de caracter칤sticas generado por la capa de convoluci칩n. Funciona tomando un subconjunto de valores (por ejemplo, el valor m치ximo o promedio) dentro de una ventana deslizante, lo que disminuye el tama침o de la representaci칩n. Esto ayuda a preservar las caracter칤sticas m치s importantes mientras se descartan detalles menos relevantes.
La reducci칩n en la dimensionalidad tambi칠n mejora la eficiencia computacional y ayuda a prevenir el sobreajuste. En resumen, el pooling simplifica la representaci칩n y permite que el modelo se enfoque en las caracter칤sticas m치s significativas.

###### C치lculo de la Dimensionalidad:
\begin{aligned}
d' = \frac{d - k}{s} + 1
\end{aligned}

Par치metros:
- **d -> dimension_in**: Es el tama침o de la dimension de entrada
- **k -> kernel_size**: Es el tama침o del kernel (matriz de orden $k^2$).
- **s -> stride**: El n칰mero de p칤xeles que el filtro se desplaza sobre la imagen de entrada, con un valor de 1 significando que se mueve un p칤xel a la vez.
- **d' -> dimension_out**: Es el tama침o de la dimension de salida.

##### Capas de convoluci칩n
Una capa de convoluci칩n extrae caracter칤sticas de la imagen de entrada mediante filtros que se deslizan sobre la misma. 
Aplica una operaci칩n matem치tica llamada convoluci칩n, produciendo un mapa de caracter칤sticas que resalta patrones como bordes y texturas. 
Mantiene la estructura espacial de la imagen, permitiendo que el modelo aprenda relaciones espaciales. 
Generalmente, se le aplica una funci칩n de activaci칩n (como ReLU) para introducir no linealidades.

###### C치lculo de la Dimensionalidad:
\begin{aligned}
d' = \frac{d + 2p - k}{s} + 1
\end{aligned}

Par치metros:
- **k -> kernel_size**: Es el orden de una matriz cuadrada.
- **s -> stride**: El n칰mero de p칤xeles que el filtro se desplaza sobre la imagen de entrada, con un valor de 1 significando que se mueve un p칤xel a la vez.
- **p -> padding**: El n칰mero de p칤xeles a침adidos alrededor de la imagen de entrada para mantener las dimensiones de salida. El padding ayuda a prevenir que la imagen se reduzca demasiado despu칠s de la convoluci칩n, permitiendo una mejor retenci칩n de informaci칩n en los bordes.


##### Capas Fully Connected
Una capa totalmente conectada (fully connected layer) conecta cada neurona de la capa anterior a cada neurona de la capa actual. Se utiliza para combinar las caracter칤sticas extra칤das por las capas previas y realizar la clasificaci칩n o regresi칩n. 
Cada neurona en esta capa aplica una transformaci칩n lineal a las entradas, seguida de una funci칩n de activaci칩n, lo que permite al modelo aprender patrones complejos. Las capas totalmente conectadas suelen encontrarse al final de una red neuronal, despu칠s de las capas de convoluci칩n y pooling.
En resumen, estas capas integran la informaci칩n y producen la salida final del modelo.


##### Capa Dropout
La capa `Dropout` es una t칠cnica de regularizaci칩n utilizada para prevenir el sobreajuste durante el entrenamiento de redes neuronales. En cada paso de entrenamiento, desactiva aleatoriamente un porcentaje de neuronas (definido por `DROPOUT_RATE`), lo que obliga a la red a no depender demasiado de una sola neurona. Esto mejora la generalizaci칩n del modelo, ayudando a que aprenda patrones m치s robustos. En esta CNN, se aplica despu칠s de la primera capa completamente conectada para mejorar el rendimiento en tareas de clasificaci칩n.

## Arquitectura 2

### Definici칩n

```python
class CNN2(nn.Module):
    def __init__(self):
        super(CNN2, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.inception1 = self.InceptionBlock(in_channels=32) # -> 128
        self.conv2 = nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(in_channels=128, out_channels=256, kernel_size=3, stride=1, padding=1)

        self.fc1 = nn.Linear(in_features=(256 * 16 * 16), out_features=64)
        self.fc2 = nn.Linear(in_features=64, out_features=3)

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dropout = nn.Dropout(DROPOUT_RATE)

    class InceptionBlock(nn.Module):
        def __init__(self, in_channels):
            super(CNN2.InceptionBlock, self).__init__()
            self.branch1x1 = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)

            self.branch3x3_1 = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)
            self.branch3x3_2 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=3, padding=1)

            self.branch5x5_1 = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)
            self.branch5x5_2 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=5, padding=2)

            self.branch_pool = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)

        def forward(self, x):
            branch1 = self.branch1x1(x)

            branch2 = self.branch3x3_1(x)
            branch2 = self.branch3x3_2(branch2)

            branch3 = self.branch5x5_1(x)
            branch3 = self.branch5x5_2(branch3)

            branch4 = F.max_pool2d(x, kernel_size=3, stride=1, padding=1)
            branch4 = self.branch_pool(branch4)

            outputs = [branch1, branch2, branch3, branch4]
            return torch.cat(outputs, 1) # (32 + 32 + 32 + 32 = 128)

    def forward(self, x):
        # Etapa 1
        x = self.pool(F.relu(self.conv1(x)))    # [32, 1, 128, 128] -conv-> [32, 32, 128, 128] -ReLu-> same -pooling-> [32, 32, 64, 64]
        x = self.inception1(x)                  # [32, 32, 64, 64] -incep-> [32, 128, 64, 64]
        x = self.pool(F.relu(self.conv2(x)))    # [32, 128, 64, 64] -conv-> [32, 128, 32, 32] -ReLu-> same -pooling-> [32, 128, 32, 32]
        x = self.pool(F.relu(self.conv3(x)))    # [32, 128, 32, 32] -conv-> [32, 256, 32, 32] -ReLu-> same -pooling-> [32, 256, 16, 16]
        # Etapa 2
        x = x.view(-1, 256 * 16 * 16)           # [32, 256, 16, 16] -view-> [32, 256 * 16 * 16]
        # Etapa 3
        x = F.relu(self.fc1(x))                 # [32, 256 * 16 * 16] -fc-> [32, 64]
        x = self.dropout(x)                     # Aplicar Dropout
        x = self.fc2(x)                         # [32, 64] -fc-> [32, 3]
        return x
```

```python
model_b_a2_r = CNN2()  # Modelo para datos crudos
model_b_a2_b = CNN2()  # Modelo para datos con filtro bilateral
model_b_a2_c = CNN2()  # Modelo para datos con filtro Canny
```

## Arquitectura 3

### Definici칩n

```python
class CNN3(nn.Module):
    def __init__(self):
        super(CNN3, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.inception1 = self.InceptionBlock(in_channels=32) # -> 128
        self.conv2 = nn.Conv2d(in_channels=128, out_channels=256, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(in_channels=256, out_channels=512, kernel_size=3, stride=1, padding=1)
        
        self.fc1 = nn.Linear(in_features=(512 * 8 * 8), out_features=64)
        self.fc2 = nn.Linear(in_features=64, out_features=3)

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dropout = nn.Dropout(DROPOUT_RATE)

    class InceptionBlock(nn.Module):
        def __init__(self, in_channels):
            super(CNN3.InceptionBlock, self).__init__()
            self.branch1x1 = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)

            self.branch3x3_1 = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)
            self.branch3x3_2 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=3, padding=1)

            self.branch5x5_1 = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)
            self.branch5x5_2 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=5, padding=2)

            self.branch_pool = nn.Conv2d(in_channels=in_channels, out_channels=32, kernel_size=1)

        def forward(self, x):
            branch1 = self.branch1x1(x)

            branch2 = self.branch3x3_1(x)
            branch2 = self.branch3x3_2(branch2)

            branch3 = self.branch5x5_1(x)
            branch3 = self.branch5x5_2(branch3)

            branch4 = F.max_pool2d(x, kernel_size=3, stride=1, padding=1)
            branch4 = self.branch_pool(branch4)

            outputs = [branch1, branch2, branch3, branch4]
            return torch.cat(outputs, 1) # (32 + 32 + 32 + 32 = 128)

    def forward(self, x):
        # Etapa 1
        x = self.pool(F.relu(self.conv1(x)))    # [32, 1, 128, 128] -conv-> [32, 32, 128, 128] -ReLu-> same -pooling-> [32, 32, 64, 64]
        x = self.pool(self.inception1(x))       # [32, 32, 64, 64] -incep-> [32, 128, 64, 64] -pooling-> [32, 128, 32, 32]
        x = self.pool(F.relu(self.conv2(x)))    # [32, 128, 32, 32] -conv-> [32, 256, 32, 32] -ReLu-> same -pooling-> [32, 256, 16, 16]
        x = self.pool(F.relu(self.conv3(x)))    # [32, 256, 16, 16] -conv-> [32, 512, 16, 16] -ReLu-> same -pooling-> [32, 512, 8, 8]
        # Etapa 2
        x = x.view(-1, 512 * 8 * 8)             # [32, 512, 8, 8] -view-> [32, 512 * 8 * 8]
        # Etapa 3
        x = F.relu(self.fc1(x))                 # [32, 512 * 8 * 8] -fc-> [32, 64]
        x = self.dropout(x)                     # Aplicar Dropout
        x = self.fc2(x)                         # [32, 64] -fc-> [32, 3]
        return x
```

# An치lisis

## 1. **CNN1**: 
   - Esta red comienza con una capa convolucional que tiene 32 filtros, seguida de un bloque Inception. El bloque Inception toma la entrada de 32 canales y la expande a 128 canales combinando varias convoluciones (1x1, 3x3, 5x5) y una operaci칩n de pooling.
   - Luego, tiene una segunda capa convolucional de 128 canales. 
   - Despu칠s del procesamiento convolucional, los datos se transforman mediante capas totalmente conectadas (fully connected), comenzando con una capa de 64 neuronas y finalizando con la capa de salida que tiene 3 neuronas, lo que indica que se est치 trabajando con un problema de clasificaci칩n de tres clases.
   - Las dimensiones de las im치genes se reducen progresivamente por medio de operaciones de "pooling" y "view", hasta que los datos son transformados para pasar por las capas fully connected.

### **Pros**:
  - Arquitectura simple y menos profunda, lo que la hace m치s r치pida de entrenar.
  - Menor riesgo de sobreajuste, ya que tiene menos par치metros que las otras redes.
  - Menor consumo de recursos computacionales (memoria y tiempo de entrenamiento).
### **Contras**:
  - Menor capacidad para aprender caracter칤sticas complejas, lo que puede limitar su rendimiento en problemas m치s dif칤ciles.
  - Menos filtros y capas pueden hacerla menos precisa en tareas de clasificaci칩n m치s complejas.


## 2. **CNN2**:
   - En este modelo, la arquitectura es similar a CNN1, pero incluye una tercera capa convolucional adicional. Despu칠s de la segunda capa convolucional con 128 filtros, CNN2 aplica una tercera convoluci칩n que incrementa los canales a 256.
   - La primera parte de la red sigue siendo casi id칠ntica: una capa convolucional, un bloque Inception y una segunda convoluci칩n. Sin embargo, la tercera convoluci칩n aumenta la capacidad del modelo para extraer caracter칤sticas m치s complejas.
   - Al final, la red transforma los datos con una capa fully connected de 64 neuronas, seguida de la capa de salida de 3 neuronas.

### **Pros**:
  - Mayor capacidad que CNN1 gracias a la adici칩n de una tercera capa convolucional con m치s filtros.
  - Puede aprender caracter칤sticas m치s complejas, lo que mejora su desempe침o en problemas m치s dif칤ciles.
  - Aumenta la complejidad sin ser excesivamente pesada en t칠rminos de c칩mputo.
### **Contras**:
  - Mayor riesgo de sobreajuste comparado con CNN1, ya que tiene m치s par치metros.
  - Consume m치s recursos (memoria y tiempo de entrenamiento) que CNN1.


## 3. **CNN3**:
   - Este modelo es el m치s profundo de los tres. Tras la capa convolucional inicial y el bloque Inception, tiene una segunda convoluci칩n que aumenta los canales a 256 (similar a CNN2), pero adem치s, a침ade una tercera capa convolucional que incrementa los canales a 512.
   - El tama침o de las caracter칤sticas se reduce progresivamente con cada operaci칩n de "pooling", y despu칠s de la tercera convoluci칩n, las dimensiones son mucho menores (8x8), lo que implica que CNN3 est치 preparada para aprender caracter칤sticas m치s abstractas.
   - Finalmente, la red tiene las capas fully connected y de salida, similares a CNN2, pero con una mayor cantidad de informaci칩n procesada en las etapas previas.

### **Pros**:
  - La arquitectura m치s profunda y con m치s filtros la hace m치s poderosa para capturar caracter칤sticas abstractas y complejas.
  - Potencialmente mejor desempe침o en tareas con patrones complicados o grandes cantidades de datos.
### **Contras**:
  - Alto riesgo de sobreajuste si no se cuenta con suficientes datos o t칠cnicas de regularizaci칩n.
  - Mayor demanda de recursos computacionales, con tiempos de entrenamiento m치s largos y mayor uso de memoria.
  - Puede ser innecesaria para tareas simples, donde una red menos profunda ser칤a m치s eficiente.
